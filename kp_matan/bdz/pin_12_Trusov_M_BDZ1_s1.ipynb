**1. Вычислить значения $n₀(0.01)$, $n₀(0.001)$.**

<big>$\lim\limits_{n\to \infty}(\frac{n^4}{2n^2-1}-\frac{n^4}{2n^2+1})$</big>

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from sympy.abc import n

n0 = 1
n_for = 10

x_n = [ (n**4)/(2*n**2-1) - (n**4)/(2*n**2+1) for n in range(n0, n_for+1)]
a = sp.limit( (n**4)/(2*n**2-1) - (n**4)/(2*n**2+1) , n, sp.oo)
eps = 0.01
print("Предел:", a)

plt.subplot(1,2,1)
x = [n for n in range(n0, n_for+1)]
y = x_n
plt.plot(x,y, 'ro')

y = [a-eps for n in range(n0, n_for+1)]
plt.plot(x,y, 'k-')

y = [a+eps for n in range(n0, n_for+1)]
plt.plot(x,y, 'k-')

a = sp.solve_univariate_inequality(sp.Abs( (n**4)/(2*n**2-1) - (n**4)/(2*n**2+1) - a )<eps,n,relational=False)
b = sp.solve_univariate_inequality(n>0,n,relational=False)
print("n₀ (", eps, ") = ", sp.Intersection(a,b), sep="")

plt.grid()
plt.title(r"$\lim_{n\to \infty}(\frac{n^4}{2n^2-1}-\frac{n^4}{2n^2+1})$, n₀=2", fontdict = {'fontsize' : 15}, pad=20)

#############################
n0 = 2
n_for = 10

x_n = [(n**4)/(2*n**2-1) - (n**4)/(2*n**2+1) for n in range(n0, n_for+1)]
a = sp.limit( (n**4)/(2*n**2-1) - (n**4)/(2*n**2+1) , n, sp.oo)
eps = 0.001

plt.subplot(1,2,2)
x = [n for n in range(n0, n_for+1)]
y = x_n
plt.plot(x,y, 'ro')

y = [a-eps for n in range(n0, n_for+1)]
plt.plot(x,y, 'k-')

y = [a+eps for n in range(n0, n_for+1)]
plt.plot(x,y, 'k-')

a = sp.solve_univariate_inequality(sp.Abs( (n**4)/(2*n**2-1) - (n**4)/(2*n**2+1) - a )<eps,n,relational=False)
b = sp.solve_univariate_inequality(n>0,n,relational=False)
print("n₀ (", eps, ") = ", sp.Intersection(a,b), sep="")

plt.grid()
plt.title(r"$\lim_{n\to \infty}(\frac{n^4}{2n^2-1}-\frac{n^4}{2n^2+1})$, n₀=4", fontdict = {'fontsize' : 15}, pad=20)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=None)
plt.show()
```

**6. Вычислить предел. В одной системе координат построить в окрестности точки x₀ графики исходной функции и функции <big>$g(x)=C(x-x₀)^p$</big>, где $p$ - порядок малости, $C$ = <big>$\lim\limits_{x\to x₀}(\frac{f(x)}{(x-x₀)^p})$</big>**

<big>$f(x) = sin(\sqrt{1+x}-1)$ относительно $x$, $ {x\to0}$</big>

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from sympy.abc import n

n0 = -1
n_for = 2

x_n = [ sp.sin(sp.sqrt(1+n)-1) for n in np.arange(n0, n_for, 0.05)]
x = [n for n in np.arange(n0, n_for, 0.05)]
y = x_n
plt.plot(x,y, 'r', label="$f(x) = sin(\sqrt{1+x}-1)$")

poryadok = 1
print("Порядок малости:", poryadok)

C = sp.limit( (sp.sin(sp.sqrt(1+n)-1))/n**poryadok , n, 0)
print("C:", C)

x_n = [ C*((n-0)**poryadok) for n in np.arange(n0, n_for, 0.05)]
x = [n for n in np.arange(n0, n_for, 0.05)]
y = x_n
plt.plot(x,y, 'b', label="$g(x)=C(x-x₀)^p$")

plt.grid()
plt.legend()

plt.show()
```

<big>$\lim\limits_{x\to 0}\frac{cos(5x)-cos(10x)}{e^{sin (x^2)}-1}$</big>

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from sympy.abc import n

a = sp.limit( (sp.cos(5*n)-sp.cos(10*n))/(np.e**(sp.sin(n))**2-1) , n, 0)
print("Предел:", a)

plt.subplot(1,2,1)
n0 = -0.05
n_for = 0.05

x_n = [ (sp.cos(5*n)-sp.cos(10*n)) for n in np.arange(n0, n_for, 0.0005)]
x = np.arange(n0, n_for, 0.0005)
y = x_n
plt.plot(x,y, 'r', label='cos(5x)-cos(10x)')

x_n = [ 75*n**2 for n in np.arange(n0, n_for, 0.0005)]
x = np.arange(n0, n_for, 0.0005)
y = x_n
plt.plot(x,y, 'b', label='$75x^2$')

plt.grid()
plt.legend()

plt.subplot(1,2,2)
n0 = -0.2
n_for = 0.2

x_n = [ np.e**(sp.sin(n**2))-1 for n in np.arange(n0, n_for, 0.0005)]
x = np.arange(n0, n_for, 0.0005)
y = x_n
plt.plot(x,y, 'r', label='$e^{\sin({x}^2)}-1$')

x_n = [ 2*(sp.sin(n**2)) for n in np.arange(n0, n_for, 0.0005)]
x = np.arange(n0, n_for, 0.0005)
y = x_n
plt.plot(x,y, 'b', label='$2\sin({x}^2)$')

plt.grid()
plt.legend()

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=None)
plt.show()
```

<big>$\lim\limits_{x\to \frac{\pi}{2}}(1+ctg(x))^{tg(x)}$</big>

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from sympy.abc import n

a = sp.limit( (1+sp.cot(n))**sp.tan(n) , n, sp.pi/2)
print("Предел:", a)

n0 = 0
n_for = 2

x_n = [ (1+sp.cot(n))**sp.tan(n) for n in np.arange(n0, n_for, 0.0005)]
x = np.arange(n0, n_for, 0.0005)
y = x_n
plt.plot(x,y, 'r', label='$(1+ctg(x))^{tg(x)}$')

x_n = [ np.e**sp.cos(n) for n in np.arange(n0, n_for, 0.0005)]
x = np.arange(n0, n_for, 0.0005)
y = x_n
plt.plot(x,y, 'b', label='${e^{\cos({x})}}$')

y = [a for n in np.arange(n0, n_for, 0.0005)]
plt.plot(x,y, 'b:', label='$a$')

plt.grid()
plt.legend()

plt.show()
```
