**1. При найденном значении <big>$a$</big> построить 2 графика в окрестности точки из условия: 1) график <big>$f(x)$</big>; 2) график <big>$f'(x)$</big>.**

<big>**$\begin{equation*}f(x)= 
 \begin{cases}
   x^2+a, x<1\\
   \frac{1}{6}(x+5), x≥1
 \end{cases}
\end{equation*}$**</big>

```python
def f1(x, a):
    return (x ** 2) + a

def f2(x):
    return 1 / 6 * (x + 5)

a = 0

plt.figure(figsize=(10, 10))

# построение функции
x = np.linspace(-5, 1, 50000)
plt.plot(x, f1(x, a), "r", label="$x^2+a$")
x = np.linspace(1, 5, 50000)
plt.plot(x, f2(x), "b", label="$(1/6)(x+5)$")

# вычисление производной средствами sympy
x = sp.Symbol('x')

y = ((x ** 2) + a)
y_der = y.diff(x)
print("f1'(x) =", y_der)

y = (1 / 6 * (x + 5))
y_der = y.diff(x)
print("f2'(x) =", y_der)

# построение производной функции
x = np.linspace(-5, 1, 50000)
plt.plot(x, 2 * x, "g-.", label="$f'(x)$")
x = np.linspace(1, 5, 50000)
plt.plot(x, 0.166666666666667 + 0 * x, "g-.")

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlim(-5, 5)
plt.ylim(-2, 5)
plt.title(r"График функции и её производной при найденном значении $a$", fontdict={'fontsize': 15}, pad=20)
plt.legend(fontsize=20)

plt.show()
```

**2. Найти производную средствами sympy, построить графики функции и её производной.**

 <big>$f(x) = (arcctg^2(x)+1)^{sin^4(4x)}$</big>

```python
plt.figure(figsize=(10, 10))

# вычисление производной средствами sympy
x = sp.Symbol('x')
y = (sp.atan(x) ** 2 + 1) ** (sp.sin(4 * x) ** 4)
y_der = y.diff(x)
print("f'(x) =", y_der)

# построение графика функции (красный цвет)
x_plt = np.arange(-4, 4, 0.1)
y_plt = []
for i in range(len(x_plt)):
    y_plt.append(y.subs(x, x_plt[i]))
plt.plot(x_plt, y_plt, "r", label="$f(x)$")

# построение графика производной функции (синий пунктир)
x_plt = np.arange(-4, 4, 0.1)
y_plt = []
for i in range(len(x_plt)):
    y_plt.append(y_der.subs(x, x_plt[i]))
plt.plot(x_plt, y_plt, "b:", label="$f'(x)$")

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.title(r"График функции $(arcctg^2(x)+1)^{sin^4(4x)}$ и её производной", fontdict={'fontsize': 15}, pad=20)
plt.legend(fontsize=20)

plt.show()
```

 <big>**$\begin{equation*} 
 \begin{cases}
   x=t^4\\
   y=e^{2t}
 \end{cases}
\end{equation*}$**</big>

**6. Найти выражение для производной средствами Python и решить уравнение вида <big>$f'(x)$</big> с помощью solve, определить тип точки экстремума (минимум/максимум).**

**Найти кратчайшее расстояние между параболой <big>$y=x^2$</big> и прямой <big>$x-y-2=0$</big>.**

```python
def f(x):
    return abs((x - x ** 2 - 2) / (np.sqrt(2)))

def f_1(x):
    return x ** 2

def f_2(x):
    return x - 2

def checkExtremum(f, x):
    eps = 0.001
    if f(x) - f(x - eps) > 0:
        return "Максимум функции"
    return "Минимум функции"

# вычисление производной средствами sympy
x = sp.Symbol('x')
y = (x - x ** 2 - 2) / (sp.sqrt(2))
y_der = y.diff(x)

print("f'(x) =", y_der)
print("f'(x) = 0 ⇔ x =", *sp.solve(y_der, x), "⇒", checkExtremum(f, sp.solve(y_der, x)[0]))
print("\nКратчайшее расстояние =", f(sp.solve(y_der, x)[0]))

plt.figure(figsize=(10, 10))
# построение функции
x = np.linspace(-5, 5, 5000)
plt.plot(x, f_1(x), "r", label="$f(x) = x^2$")
x = np.linspace(-5, 5, 5000)
plt.plot(x, f_2(x), "b", label="$x-y-2=0$")
x = np.linspace(-5, 5, 5000)
plt.plot(x, -x + 0.75, "g", label="$y=-x+0.75$")

plt.scatter(1 / 2, f_1(1 / 2), s=100, c="red")
plt.scatter(11 / 8, f_2(11 / 8), s=100, c="blue")

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title(r"Кратчайшее расстояние между параболой $y=x^2$ и прямой $x-y-2=0$", fontdict={'fontsize': 15}, pad=20)
plt.legend(fontsize=20)

plt.show()
```

```python
def f(x):
    return np.log(x ** 2 - 1) / (x ** 2 - 1)

def der(x_podst):
    y = -2 * x * sp.log(x ** 2 - 1) / (x ** 2 - 1) ** 2 + 2 * x / (x ** 2 - 1) ** 2
    return y.subs(x, x_podst)

def checkExtremum(f, x):
    eps = 0.001
    if f(x) - f(x - eps) > 0:
        return "Максимум"
    return "Минимум"

plt.figure(figsize=(10, 10))

# построение функции
x = np.linspace(-7, -1.001, 5000)
plt.plot(x, f(x), "r", label="$f(x)$")
x = np.linspace(1.001, 7, 5000)
plt.plot(x, f(x), "r")

# построение производной функции
x = sp.Symbol('x')
x_plt = np.arange(-7, -1, 0.01)
y_plt = []
for i in range(len(x_plt)):
    y_plt.append(der(x_plt[i]))
plt.plot(x_plt, y_plt, "b:", label="$f'(x)$")

x_plt = np.arange(1.01, 7, 0.01)
y_plt = []
for i in range(len(x_plt)):
    y_plt.append(der(x_plt[i]))
plt.plot(x_plt, y_plt, "b:")

# точки экстремума
x1 = optimize.brentq(der, -3, -1.1)
print("Точка экстремума 1:", x1, "-", checkExtremum(f, x1), "Значение:", f(x1))
x2 = optimize.brentq(der, 1.1, 3)
print("Точка экстремума 2:", x2, "-", checkExtremum(f, x2), "Значение:", f(x2))

# построение точек экстремума
plt.scatter(x1, f(x1), s=100, c="red")
plt.scatter(x2, f(x2), s=100, c="red")

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlim(-7, 7)
plt.ylim(-2, 2)
plt.title(r"Точки экстремума функции $y=\frac{ln(x^2-1)}{x^2-1}$", fontdict={'fontsize': 15}, pad=20)
plt.legend(fontsize=20)

plt.show()
```
