```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import math
```

**Упражнение 1. Изобразить область интегрирования. Вычислить интеграл, расставив пределы интегрирования двумя способами:**

**<big>$\iint\limits_{G}(x+y^2)dxdy$</big>, где <big>$G$</big> ограничена кривыми <big>$y=2x$</big> и <big>$y=x^2$</big>.**

```python
def solid_solid(f, g, a, b, n):
    plt.figure(figsize=(10, 10))
    x = np.linspace(a, b, 10000)
    plt.plot(x, f(x), "b-")
    plt.plot(x, g(x), "b-")
    dx = (b - a) / n
    c = a
    while c <= b:
        plt.vlines(x=c, ymin=f(c), ymax=g(c), color="b")
        c += dx
    # параметры координатной плоскости
    plt.xlabel("x")
    plt.ylabel("y", rotation=0)
    plt.grid()
    plt.axhline(y=0, color="k")
    plt.axvline(x=0, color="k")
    plt.show()

def f(x):
    return 2 * x

def g(x):
    return x**2

solid_solid(f, g, 0, 2, 24)
```

**<big>$\iint\limits_{G}(x+y^2)dxdy = \int_0^2 dx\int_{x^2}^{2x} (x+y^2)dy = \int_0^4 dy\int_{\frac{y}{2}}^{\sqrt{y}} (x+y^2)dx$</big>**

```python
x, y = sp.symbols("x y")
f = x + y**2

a, b = (0, 2)
c, d = (x**2, 2 * x)
I_in = sp.integrate(f, (y, c, d))
I_out = sp.integrate(I_in, (x, a, b))
print(f"I1 = {I_out}")

a, b = (0, 4)
c, d = (y / 2, sp.sqrt(y))
I_in = sp.integrate(f, (x, c, d))
I_out = sp.integrate(I_in, (y, a, b))
print(f"I2 = {I_out}")
```

**Упражнение 2. Изобразить область интегрирования. Вычислить тройной интеграл <big>$\iiint\limits_{V}(z)dxdydz$</big>, где <big>$V$</big> ограничена координатными плоскостями <big>$x=0$</big>, <big>$y=0$</big>, <big>$z=0$</big> и плоскостью <big>$x+y+z=1$</big>.**

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(projection="3d")

xs = np.linspace(0, 1, 50)
ys = np.linspace(0, 1, 50)
X, Y = np.meshgrid(xs, ys)
Z = 0 + 0 * X + 0 * Y
surf = ax.plot_surface(X, Y, Z, color="white", linewidth=0)

ys = np.linspace(0, 1, 50)
zs = np.linspace(0, 1, 50)
Y, Z = np.meshgrid(ys, zs)
X = 0 + 0 * Y + 0 * Z
surf = ax.plot_surface(X, Y, Z, color="white", linewidth=0)

xs = np.linspace(0, 1, 50)
zs = np.linspace(0, 1, 50)
X, Z = np.meshgrid(xs, zs)
Y = 0 + 0 * X + 0 * Z
surf = ax.plot_surface(X, Y, Z, color="white", linewidth=0)

xs = np.linspace(0, 1, 50)
ys = np.linspace(0, 1, 50)
X, Y = np.meshgrid(xs, ys)
Z = 1 - X - Y
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.magma, linewidth=0)

# параметры координатной плоскости
ax.elev = 45
ax.azim = 30
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_xlim(0, 1.5)
ax.set_ylim(0, 1.5)
ax.set_zlim(0, 1.5)
plt.grid()

plt.show()
```

**<big>$\iiint\limits_{V}(z)dxdydz = \int_0^1 dx\int_{0}^{1-x}dy\int_{0}^{1-x-y}(z)dz$</big>**

```python
x, y, z = sp.symbols("x y z")
f = z

a, b = (0, 1)
c, d = (0, 1 - x)
e, e1 = (0, 1 - x - y)
I_in = sp.integrate(f, (z, e, e1))
I_center = sp.integrate(I_in, (y, c, d))
I_out = sp.integrate(I_center, (x, a, b))
print(f"I = {I_out}")
```

**Упражнение 3. Изобразить область интегрирования. Вычислить интеграл двумя способами.**

**<big>$\iint\limits_{S}\frac{(1-x)}{(y+1)^2}dxdy$</big> ; <big>$S = \begin{equation}
    \begin{cases}
      0 \leq x \leq 1\\
       0 \leq y \leq x^2
    \end{cases}\,
\end{equation}$</big>**

```python
def solid_solid(f, g, a, b, n):
    plt.figure(figsize=(10, 10))
    x = np.linspace(a, b, 10000)
    plt.plot(x, f(x), "b-")
    plt.plot(x, g(x), "b-")
    dx = (b - a) / n
    c = a
    while c <= b:
        plt.vlines(x=c, ymin=f(c), ymax=g(c), color="b")
        c += dx
    # параметры координатной плоскости
    plt.xlabel("x")
    plt.ylabel("y", rotation=0)
    plt.grid()
    plt.axhline(y=0, color="k")
    plt.axvline(x=0, color="k")
    plt.show()

def f(x):
    return 0*x

def g(x):
    return x**2

solid_solid(f, g, 0, 1, 24)
```

<big>$\iint\limits_{S}\frac{(1-x)}{(y+1)^2}dxdy = \int_0^1 dx\int_{0}^{x^2} \frac{(1-x)}{(y+1)^2}dy = \int_0^1 dy\int_{\sqrt{y}}^{1} \frac{(1-x)}{(y+1)^2}dx$</big>

```python
x, y = sp.symbols("x y")
f = (1 - x) / (y + 1) ** 2

a, b = (0, 1)
c, d = (0, x**2)
I_in = sp.integrate(f, (y, c, d))
I_out = sp.integrate(I_in, (x, a, b))
print(f"I1 = {I_out}")

a, b = (0, 1)
c, d = (sp.sqrt(y), 1)
I_in = sp.integrate(f, (x, c, d))
I_out = sp.integrate(I_in, (y, a, b))
print(f"I2 = {I_out}")
```

**!Упражнение 1C. Изобразить область интегрирования. Вычислить интеграл, расставив пределы интегрирования двумя способами:**

**<big>$\iint\limits_{G}(sin(xy))dxdy$</big>, где <big>$G$</big> ограничена кривыми <big>$y=1-2x^2$</big> и <big>$y+x=0$</big>.**

```python
def solid_solid(f, g, a, b, n):
    plt.figure(figsize=(10, 10))
    x = np.linspace(a, b, 10000)
    plt.plot(x, f(x), "b-")
    plt.plot(x, g(x), "b-")
    dx = (b - a) / n
    c = a
    while c <= b:
        plt.vlines(x=c, ymin=f(c), ymax=g(c), color="b")
        c += dx
    # параметры координатной плоскости
    plt.xlabel("x")
    plt.ylabel("y", rotation=0)
    plt.grid()
    plt.axhline(y=0, color="k")
    plt.axvline(x=0, color="k")
    plt.show()

def f(x):
    return 1 - 2 * x**2

def g(x):
    return -x

solid_solid(f, g, -0.5, 1, 24)
```

**<big>$\iint\limits_{G}(sin(xy))dxdy = \int_{-0.5}^1 dx\int_{-x}^{1-2x^2} (sin(xy))dy = \int_{-1}^{0.5} dy\int_{-y}^{\sqrt{\frac{1-y}{2}}} (sin(xy))dx + \int_{0.5}^{1} dy\int_{-\sqrt{\frac{1-y}{2}}}^{\sqrt{\frac{1-y}{2}}} (sin(xy))dx$</big>**

```python
x, y = sp.symbols("x y")
f = sp.sin(x*y)

a, b = (-0.5, 1)
c, d = (-x, 1-2*x**2)
I_in = sp.integrate(f, (y, c, d))
I_out = sp.integrate(I_in, (x, a, b))
print(f"I1 = {I_out}")

a, b = (-1, 0.5)
c, d = (-y, -sp.sqrt((1-y)/2))
I_in = sp.integrate(f, (x, c, d))
I_out = sp.integrate(I_in, (y, a, b))

a, b = (0.5, 1)
c, d = (-sp.sqrt((1-y)/2), sp.sqrt((1-y)/2))
I_in = sp.integrate(f, (x, c, d))
I_out2 = sp.integrate(I_in, (y, a, b))

print(f"I2 = {(I_out + I_out2)}")
```

**Упражнение C2. Изобразить область интегрирования. Вычислить тройной интеграл <big>$\iiint\limits_{V}(xyz)dxdydz$</big>, где <big>$V$</big> ограничена координатными плоскостями <big>$x=y^2$</big>, <big>$y=x^2$</big>, <big>$z=xy$</big> и плоскостью <big>$z=0$</big>.**

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(projection="3d")

xs = np.linspace(0, 1, 50)
ys = np.linspace(0, 1, 50)
X, Y = np.meshgrid(xs, ys)
Z = 0 + 0 * X + 0 * Y
surf = ax.plot_surface(X, Y, Z, color="white", linewidth=0)

ys = np.linspace(0, 1, 50)
zs = np.linspace(0, 1, 50)
Y, Z = np.meshgrid(ys, zs)
X = Y**2
surf = ax.plot_surface(X, Y, Z, color="white", linewidth=0)

xs = np.linspace(0, 1, 50)
zs = np.linspace(0, 1, 50)
X, Z = np.meshgrid(xs, zs)
Y = X**2
surf = ax.plot_surface(X, Y, Z, color="white", linewidth=0)

xs = np.linspace(0.5, 1, 50)
ys = np.linspace(0.5, 1, 50)
X, Y = np.meshgrid(xs, ys)
Z = 1 / (X * Y)
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.magma, linewidth=0)

# параметры координатной плоскости
ax.elev = 15
ax.azim = 15
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.grid()

plt.show()
```

**<big>$\iiint\limits_{V}(xyz)dxdydz = \int_0^1 dx\int_{0}^{x^2}dy\int_{0}^{xy}(xyz)dz$</big>**

```python
x, y, z = sp.symbols("x y z")
f = x * y * z

a, b = (0, 1)
c, d = (0, x**2)
e, e1 = (0, x * y)
I_in = sp.integrate(f, (z, e, e1))
I_center = sp.integrate(I_in, (y, c, d))
I_out = sp.integrate(I_center, (x, a, b))
print(f"I = {I_out}")
```
