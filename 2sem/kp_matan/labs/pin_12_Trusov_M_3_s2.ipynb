**Упражнение 1. Создайте функцию, которая строит график последовательности частичных сумм ряда. В качестве входных параметров функции использовать формулу <big>$a_n$</big> общего члена последовательности и число <big>$n_0$</big> рассматриваемых членов.**

```python
def drawSum(a_n, n):
    csums = [a_n.subs(x, 1)]
    for i in range(2, n + 1):
        csums.append((csums[i - 2]) + a_n.subs(x, i))

    plt.plot([i for i in range(1, n + 1)], csums, "r")

    # параметры координатной плоскости
    plt.xlabel("x")
    plt.ylabel("y", rotation=0)
    plt.grid()
    plt.axhline(y=0, color="k")
    plt.axvline(x=0, color="k")

    plt.show()
```

**Упражнение 2.**

```python
def getShod(a_n):
    x = sp.Symbol("x")
    lim = sp.limit(a_n, x, sp.oo)
    if lim==sp.oo:
        return "Ряд расходится"
    else:
        I = sp.integrate(a_n, (x, 1, sp.oo))
        if I == sp.oo:
            return "Ряд расходится"
    return "Ряд сходится"
```

```python
def isBigger(a_n, b_n):
    for i in range(1, 500):
        if a_n.subs(x, i) < b_n.subs(x, i):
            return False
    return True
```

```python
def drawBoth(a_n, b_n, n):

    plt.subplot(1, 2, 1)
    posls = []
    for i in range(1, n + 1):
        posls.append(a_n.subs(x, i))
    plt.plot([i for i in range(1, n + 1)], posls, "r")

    posls = []
    for i in range(1, n + 1):
        posls.append(b_n.subs(x, i))
    plt.plot([i for i in range(1, n + 1)], posls, "b:")
    
    # параметры координатной плоскости
    plt.xlabel("x")
    plt.ylabel("y", rotation=0)
    plt.grid()
    plt.axhline(y=0, color="k")
    plt.axvline(x=0, color="k")

    plt.subplot(1, 2, 2)
    csums = [a_n.subs(x, 1)]
    for i in range(2, n + 1):
        csums.append((csums[i - 2]) + a_n.subs(x, i))

    plt.plot([i for i in range(1, n + 1)], csums, "r")

    csums = [b_n.subs(x, 1)]
    for i in range(2, n + 1):
        csums.append((csums[i - 2]) + b_n.subs(x, i))

    plt.plot([i for i in range(1, n + 1)], csums, "b:")

    # параметры координатной плоскости
    plt.xlabel("x")
    plt.ylabel("y", rotation=0)
    plt.grid()
    plt.axhline(y=0, color="k")
    plt.axvline(x=0, color="k")

    plt.show()
```

**б) Геометрически проиллюстрируйте использование признака сравнения: для каждой пары сравниваемых рядов постройте в одной системе координат графики последовательностей общих членов, а в другой - графики последовательностей их частичных сумм.**

```python
x = sp.Symbol("x")
first = 1 / (x * sp.sqrt(x + 1) * sp.sqrt(x + 2))
first_1 = 1 / x**2

print(isBigger(first_1, first))
print(getShod(first_1))
drawBoth(first, first_1, 25)
```

**Упражнение 7. Пусть к ряду <big>$\sum_{n=1}^{\infty} {a_n}$</big> применимо утверждение об оценке ряда. Создайте функцию, которая оценивает число членов, достаточное для вычисления суммы ряда с заданной точностью <big>$ε$</big>, вычисляет сумму ряда с заданной точностью. В качестве входных параметров функции используйте формулу общего члена последовательности и точность <big>$ε$</big>.**

```python
def countSeries(a_n, eps):
    c = 1
    summa = 0
    while abs(a_n.subs(x, c)) > eps:
        summa += a_n.subs(x, c)
        c += 1
    return c, float(summa)
```
