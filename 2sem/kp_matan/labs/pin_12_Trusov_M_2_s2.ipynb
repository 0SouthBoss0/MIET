```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

**Упражнение 1. Построить график функции <big>$y=xcos(x)$</big> на отрезке <big>$[0;\frac{\pi}{2}]$</big>. Вычислить площадь фигуры, ограниченной графиком функции и линиями <big>$x=0$</big>, <big>$x=\frac{π}{2}$</big>, <big>$y=0$</big>.**

<big>$S=\int_{a}^{b} f(x) \,dx$</big>

```python
x_plt = np.linspace(0, np.pi / 2, 50000)

plt.plot(x_plt, x_plt * np.cos(x_plt), "r")
plt.axvline(x=np.pi / 2, color="k", linestyle=":")

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlim(-0.5, 2)
plt.ylim(-0.5, 1)

plt.show()
```

```python
x = sp.Symbol("x")

I = sp.integrate(x * sp.cos(x), (x, 0, sp.pi / 2))
sp.pprint(I)
```

**Упражнение 2. Построить графики функций <big>$y=x^2+2x$</big> и <big>$y = 7-4x-x^2$</big>. Найти точки пересечения графиков. Вычислить площадь фигуры, ограниченной графиками.**

<big>$S=\int_{a}^{b} (f(x)-g(x)) \,dx$</big>

```python
x_plt = np.linspace(-10, 10, 50000)

plt.plot(x_plt, x_plt**2 + 2*x_plt, "r", label="$y=x^2+2x$")
plt.plot(x_plt, 7-4*x_plt-x_plt**2, "b", label="$y=7-4x-x^2$")

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlim(-7, 5)
plt.ylim(-5, 15)

plt.legend()
plt.show()
```

```python
x = sp.Symbol("x")

ans = sp.solve(x**2 + 2 * x - (7 - 4 * x - x**2), x)
ans.sort()

print("Точки пересечения графиков:")
for i in range(len(ans)):
    print(ans[i])
```

```python
x = sp.Symbol("x")

I = sp.integrate(7 - 4 * x - x**2 - (x**2 + 2 * x), (x, ans[0], ans[1]))
print(float(sp.simplify(I)))
```

**Упражнение 3. Построить графики функций, заданные параметрически. Вычислить площадь фигуры, ограниченной графиками.**

<big>$S=\int_{t_1}^{t_2} y(t)x'(t)\,dt$</big>

**а) <big>$x=cos(t)$</big>, <big>$y=sin(t)$</big>, <big>$t∈[0;2\pi]$</big>;**

```python
t_plt = np.linspace(0, 2 * np.pi, 100)

x_plt = np.cos(t_plt)
y_plt = np.sin(t_plt)
plt.plot(x_plt, y_plt)

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k")

plt.show()
```

```python
t = sp.Symbol("t")

I = sp.integrate(sp.sin(t) * (sp.cos(t).diff(t)), (t, 2 * sp.pi, 0))
sp.pprint(I)
```

**б) <big>$x=2cos(t)$</big>, <big>$y=3sin(t)$</big>, <big>$t∈[0;2\pi]$</big>.**

```python
t_plt = np.linspace(0, 2 * np.pi, 100)

x_plt = 2 * np.cos(t_plt)
y_plt = 3 * np.sin(t_plt)
plt.plot(x_plt, y_plt)

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k")

plt.show()
```

```python
t = sp.Symbol("t")

I = sp.integrate(3 * sp.sin(t) * (2 * sp.cos(t).diff(t)), (t, 2 * sp.pi, 0))
sp.pprint(I)
```

**Упражнение 4. Построить фигуру, ограниченную графиком логарифмической спирали <big>$r = e^φ$</big> и прямыми <big>$φ=0$</big>, <big>$φ=2\pi$</big>. Найти площадь фигуры.**

<big>$S=\frac{1}{2}\int_{\alpha}^{\beta} r^2(φ)\,dφ$</big>

```python
plt.axes(projection="polar")

phi_plt = np.linspace(0, 2 * np.pi, 1000)
plt.polar(phi_plt, np.exp(phi_plt), "r")

r_plt = np.linspace(0, np.exp(2 * np.pi), 1000)
plt.polar(0 + 0 * r_plt, r_plt, "b")
plt.polar(2 * sp.pi + 0 * r_plt, r_plt, "b")

plt.show()
```

```python
φ = sp.Symbol("φ")

I = 0.5 * sp.integrate((sp.exp(φ)) ** 2, (φ, 0, 2 * sp.pi))
sp.pprint(I)
```

**Упражнение 5. Найти длину дуги параболы <big>$y = x^2$</big>, от точки <big>$A(1; 1)$</big> до точки <big>$B(2;4)$</big>.**

<big>$S=\int_{a}^{b} \sqrt{1+(y'(x))^2}\,dx$</big>

```python
x_plt = np.linspace(-1, 5, 50000)
plt.plot(x_plt, x_plt**2, "b:")

x_plt = np.linspace(1, 2, 50000)
plt.plot(x_plt, x_plt**2, "r")

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k")
plt.xlim(-1, 3)
plt.ylim(-1, 5)

plt.show()
```

```python
x = sp.Symbol("x")

I = sp.integrate(sp.sqrt(1 + ((x**2).diff(x)) ** 2), (x, 1, 2))
print(float(I))
```

**Упражнение 6. Найти длину замкнутой кривой, заданной параметрическими уравнениями <big>$x=2cos(t)$</big>, <big>$y=3sin(t)$</big>, <big>$t∈[0;2\pi]$</big>.**

<big>$S=\int_{t_1}^{t_2} \sqrt{(x'(t))^2+(y'(t))^2}\,dt$</big>

```python
t_plt = np.linspace(0, 2 * np.pi, 100)

x_plt = 2 * np.cos(t_plt)
y_plt = 3 * np.sin(t_plt)
plt.plot(x_plt, y_plt)

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k")

plt.show()
```

```python
t = sp.Symbol("t")

I = sp.integrate(sp.sqrt(((2 * sp.cos(t).diff(t)) ** 2) + ((3 * sp.sin(t).diff(t)) ** 2)), (t, 0, 2 * sp.pi))
sp.pprint(float(I))
```

**Упражнение 7. Вычислить длину замкнутой кривой, задаваемой уравнением <big>$r=4(1+cos(φ))$</big>.**

<big>$S=\int_{\alpha}^{\beta} \sqrt{(r(φ))^2+(r'(φ))^2}\,dφ$</big>

```python
plt.axes(projection="polar")
phi_plt = np.linspace(0, 2 * np.pi, 1000)
plt.polar(phi_plt, 4 * (1 + np.cos(phi_plt)), "r")

plt.show()
```

```python
φ = sp.Symbol("φ")

I = sp.integrate(sp.sqrt((4 * (1 + sp.cos(φ))) ** 2 + ((4 * (1 + sp.cos(φ))).diff(φ)) ** 2), (φ, 0, 2 * sp.pi))
sp.pprint(float(I))
```

**Упражнение 8. Вычислить объем тела, полученного при вращении криволинейной трапеции, ограниченной графиком функции <big>$y=sin(x)$</big> и прямой $y=0$ <big>$(x∈[0;\pi])$</big>.**

**а) относительно оси $Ox$;**

<big>$V=\pi\int_{a}^{b} y^2(x)\,dx$</big>

```python
x = sp.Symbol("x")

I = sp.pi * sp.integrate(sp.sin(x) ** 2, (x, 0, sp.pi))
sp.pprint(I)
```

**б) относительно оси $Oy$.**

<big>$V=2\pi\int_{a}^{b} xy(x)\,dx$</big>

```python
x = sp.Symbol("x")

I = 2 * sp.pi * sp.integrate(x * sp.sin(x), (x, 0, sp.pi))
sp.pprint(I)
```

**Упражнение 1С. Построить графики функций <big>$y=x^2+1$</big>, <big>$y=3-x$</big>, <big>$y=1$</big>. Вычислить площадь фигуры, ограниченной графиками всех трех функций. График оформить: сделать одинаковый масштаб по осям, нанести сетку, пометить оси координат, сделать заголовок.**

```python
plt.figure(figsize=(10, 10))

x_plt = np.linspace(-10, 10, 50000)
plt.plot(x_plt, x_plt**2 + 1, "r", label="$y_1=x^2+1$")
plt.plot(x_plt, 3 - x_plt, "g", label="$y_2=3-x$")
plt.plot(x_plt, 1 + 0 * x_plt, color="b", label="$y_3=1$")

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.grid()
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k")
plt.xlim(-4, 4)
plt.ylim(-1, 7)
plt.title(r"Упражнение $1C$", fontdict={"fontsize": 20}, pad=20)
plt.legend(fontsize=15)

plt.show()
```

<big>$S=\int_{0}^{1} (y_1(x)-y_3(x)) \,dx + \int_{1}^{2} (y_2(x)-y_3(x))$</big>

```python
x = sp.Symbol("x")

I = sp.integrate((x**2 + 1) - 1, (x, 0, 1)) + sp.integrate((3 - x) - 1, (x, 1, 2))
print(I)
```
