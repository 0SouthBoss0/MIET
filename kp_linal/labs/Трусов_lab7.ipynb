**Упражнение 7.1. Построить эллипс с центром в точке $(3,1)$, горизонтальной полуосью, равной $3$, и эксцентриситетом, равным $4/5$. Нанести на рисунок центр эллипса и его фокусы. Найти все параметры эллипса и его параметрическое и каноническое уравнения.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

e = Ellipse(Point(3, 1), hradius=3, eccentricity=Rational(4, 5))

print("Центр эллипса:", e.center)
print("Величина горизонтальной полуоси эллипса (a):", e.hradius)
print("Величина вертикальной полуоси эллипса (b):", e.vradius)
print("Площадь эллипса:", e.area)
print("Эксцентриситет эллипса:", e.eccentricity)
print("Фокусы эллипса:", *e.foci)
print("Расстояние от центра до фокуса эллипса (c):", e.focus_distance)
print("Каноническое уравнение эллипса:", e.equation())
print("Параметрическое уравнение эллипса:", *e.arbitrary_point())
```

```python
# построение фокусов эллипса
plt.plot(e.foci[0][0], e.foci[0][1],'ro')
plt.plot(e.foci[1][0], e.foci[1][1],'ro')

# построение центра эллипса
plt.plot(x_center, y_center, 'bo')

# построение эллипса
t = np.arange(0, 2*np.pi, 0.01)
x = e.hradius*np.cos(t)
y = e.vradius*np.sin(t)
plt.plot(e.center[0]+x, e.center[1]+y, lw=2)

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation = 0)
plt.grid()
plt.axis("equal")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

**Упражнение 7.2. Построить параболу с фокусом в точке $(0,0)$ и директрисой $y=8$. Нанести на рисунок фокус параболы и директрису. Найти все параметры и каноническое уравнение.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

p = Parabola(Point(0, 0), Line(Point(0, 8), Point(1, 8)))

print("Параметр параболы (p):", p.p_parameter)
print("Директриса параболы:", p.directrix.equation())
print("Ось симметрии параболы:", p.axis_of_symmetry.equation())
print("Фокус параболы:", *p.focus)
print("Расстояние от центра до фокуса параболы (c):", p.focal_length)
print("Эксцентриситет параболы:", p.eccentricity)
print("Каноническое уравнение параболы:", p.equation())
```

```python
#построение фокуса параболы
plt.plot(p.focus[0], p.focus[1],'ro')

#построение директрисы параболы
plt.axhline(y=8, color='r')

# построение параболы
x = np.arange(-20, 20, 0.01)
y = (-x**2+64)/16
plt.plot(x, y, lw=2,color='b')

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation = 0)
plt.grid()
plt.axis("equal")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

**Упражнение 7.3. Построить кривую <big>$\frac{y^2}{2}-\frac{(x-2)^2}{8}=1 $</big>, ее асимптоты и фокусы.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

a = np.sqrt(2)
b = np.sqrt(8)
x_center = 2
y_center = 0
c = np.sqrt(a**2+b**2)

# построение гиперболы
x = np.arange(-10, 10, 0.01)
y1 = np.sqrt((8+(x-2)**2)/4)
y2 = -np.sqrt((8+(x-2)**2)/4)
plt.plot(x, y1, lw=2,color='b')
plt.plot(x, y2, lw=2,color='b')

# построение асимптот
x = np.arange(-10, 10, 0.01)
y1 = (a/b)*(x-x_center) + y_center
y2 = -(a/b)*(x-x_center) + y_center
plt.plot(x, y1, "k:")
plt.plot(x, y2, "k:")

#построение фокусов гиперболы
plt.plot(x_center, y_center - c,'ro')
plt.plot(x_center, y_center + c,'ro')

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation = 0)
plt.grid()
plt.axis("equal")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

**Упражнение 7.4. Найти параметрический вид кривой <big>$\frac{y^2}{2}-\frac{(x-2)^2}{8}=1 $</big>. Построить эту кривую (возьмите <big>$t∈[-3;3]$</big>). Показать, как меняются координаты точки в зависимости от параметра t.**

```python
import math

a = np.sqrt(2)
b = np.sqrt(8)
x_center = 2
y_center = 0

# построение гиперболы
t = np.arange(-3, 3, 0.01)

# x = a * ch(t)
x = a * (np.e**t+np.e**-t)*0.5 + y_center 
# x = -a * ch(t)
x2 = -a * (np.e**t+np.e**-t)*0.5 + y_center
# y = b * sh(t)
y = b * (np.e**t-np.e**-t)*0.5 + x_center

plt.plot(y,x, color='b')
plt.plot(y,x2, color='b')

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation = 0)
plt.grid()
plt.axis("equal")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

**Упражнение 7.5. Найти точки пересечения параболы <big>$x^2=y+1$</big> и эллипса  <big>$\frac{x^2}{4}+\frac{y^2}{25}=1$</big>. Сделать рисунок. Нанести на рисунок полученные точки.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

p = Parabola(Point(0, -3/4), Line(Point(0, -1.25), Point(1, -1.25)))
print("Каноническое уравнение параболы:", p.equation())

e = Ellipse(Point(0, 0), hradius=2, vradius=5)
print("Каноническое уравнение эллипса:", e.equation())

print("Точка пересечения 1:", p.intersection(e)[0])
print("Точка пересечения 2:", p.intersection(e)[1])
```

```python
# построение эллипса
t = np.arange(0, 2*np.pi, 0.01)
x = e.hradius*np.cos(t)
y = e.vradius*np.sin(t)
plt.plot(e.center[0]+x, e.center[1]+y, lw=2)

# построение параболы
x = np.arange(-3, 3, 0.01)
y = x**2-1
plt.plot(x, y, lw=2,color='b')

# построение точек пересечения
plt.plot(np.sqrt(-17/8 + 5*np.sqrt(73)/8), -25/8 + 5*np.sqrt(73)/8,'ro')
plt.plot(-np.sqrt(-17/8 + 5*np.sqrt(73)/8), -25/8 + 5*np.sqrt(73)/8,'ro')

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation = 0)
plt.grid()
plt.axis("equal")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```
