**Упражнение 3.1.

**Упражнение 3.3. 
Проверить являются ли векторы p, q, r линейно зависимыми.**

```python
# p = a-b+c
# q = a+b
# r = b-c

A = np.matrix ('1 -1 1; 1 1 0; 0 1 -1')
B = np.matrix ('0; 0; 0')

A_inv = np.linalg.inv(A)
print(A_inv.dot(B))
```

*Таким образом, не существует ненулевой комбинации значений чисел x1, x2, x3 таких, что линейная комбинация векторов p, q, r равна нулевому вектору.
Это значит, что вектора p, q, r линейно независимы (образуют базис).*

**Разложить вектор s по этим векторам.**

```python
# s = a+b+c

A = np.matrix ('1 -1 1; 1 1 0; 0 1 -1')
A = A.transpose()
B = np.matrix ('1; 1; 1')

A_inv = np.linalg.inv(A)
print(A_inv.dot(B))
```

*Ответ: s = -p + 2q - 2r*

**Упражнение 3.4. Даны три точки A(–2,0), B(3,4), C(4,–1). Найти величины углов треугольника АВС. Проверить свойство суммы углов треугольника. Сделать рисунок.**

```python
# координаты трех точек
OA = np.array([-2, 0])
OB = np.array([3, 4])
OC = np.array([4, -1])

# нахождение координат векторов
AB = OB - OA
BC = OC - OB
CA = OA - OC
CB = OB - OC
AC = OC - OA
BA = OA - OB

# нахождение угла между векторами
cos_angle1 = np.dot(AB, CB) / (np.linalg.norm(AB) * np.linalg.norm(CB))
alfa1 = np.arccos(cos_angle1)
alfa1 = math.degrees(alfa1)

cos_angle2 = np.dot(CA, BA) / (np.linalg.norm(CA) * np.linalg.norm(BA))
alfa2 = np.arccos(cos_angle2)
alfa2 = math.degrees(alfa2)

cos_angle3 = np.dot(AC, BC) / (np.linalg.norm(AC) * np.linalg.norm(BC))
alfa3 = np.arccos(cos_angle3)
alfa3 = math.degrees(alfa3)

print("Углы треугольника:", alfa1, alfa2, alfa3)
print("Верно ли свойство суммы углов треугольника:", alfa1+alfa2+alfa3 == 180)

# сетка координатной плоскости
plt.grid()

# установка размеров координатной плоскости
plt.ylim(-2, 5)
plt.xlim(-3, 5)

# построение векторов
plt.arrow(OA[0], OA[1], AB[0], AB[1], width=0.07, length_includes_head=True, facecolor='Yellow')
plt.arrow(OB[0], OB[1], BC[0], BC[1], width=0.07, length_includes_head=True, facecolor='Red')
plt.arrow(OC[0], OC[1], CA[0], CA[1], width=0.07, length_includes_head=True, facecolor='Blue')

# вывод изображения
plt.show()
```

**Упражнение 3.6. Вычислить площадь треугольника АВС с вершинами A(1,3,–1), B(2,–1,4), C(5,0,3). Как соотносятся площадь треугольника и модуль векторного произведения?**

```python
# координаты трех точек
OA = np.array([1, 3, -1])
OB = np.array([2, -1, 4])
OC = np.array([5, 0, 3])

# нахождение координат векторов
AC = OC - OA
AB = OB - OA
BC = OC - OB
CA = OA - OC

# нахождение векторного произведения
proizvedenie = np.cross(AC, AB)

# нахождение модуля векторного произведения
modul_proizvedenia = np.linalg.norm(proizvedenie)

# нахождение площади треугольника
S = modul_proizvedenia / 2

print("Площадь треугольника:", S)
print("Модуль векторного произведения:", modul_proizvedenia)
```

*Модуль векторного произведения двух векторов a и b равен площади параллелограмма построенного на этих векторах: S парал = [a × b].*

*Площадь треугольника построенного на векторах a и b равна половине модуля векторного произведения этих векторов.*

```python
# построение для трёхмерного пространства# построение для трёхмерного пространства
ax = plt.axes(projection='3d')

# установка размеров координатного пространства
ax.set_xlim([0, 4])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])

# построение векторов
ax.quiver(*OA, *AB, color='Blue')
ax.quiver(*OB, *BC, color='Green')
ax.quiver(*OC, *CA, color='Red')

# сетка координатного пространства
plt.grid()

# вывод изображения
plt.show()
```

**Упражнение С1. Используя определение смешанного произведения, вычислить объем пирамиды ABCD. Проверить с помощью определителя**

Смешанным произведением трёх векторов а, b, с называется скалярное произведение вектора а на векторное произведение b×с, т.е. а⋅(b×с) или (b×с)⋅а

Объём треугольной пирамиды, построенной на трех векторах равен 1/6 модуля смешанного произведения векторов.

```python
# AB = 3i+4j
# AC = -3i+k
# AD = 2j+5k

# нахождение координат векторов
AB = np.array([3, 4, 0])
AC = np.array([-3, 0, 1])
AD = np.array([0, 2, 5])

# нахождение векторного произведения векторов b*c
vector_proizvedenie = np.cross(AC, AD)

# нахождение скалярного произведения вектора a на векторное произведение b×с
scalar_proizvedenie = np.dot(AB, vector_proizvedenie)

# нахождение модуля скалярного произведения вектора a на векторное произведение b×с
modul_proizvedenie = np.linalg.norm(scalar_proizvedenie)

# нахождение объема треугольной пирамиды
V = modul_proizvedenie / 6
print(V)
```

Проверка с помощью определителя.

Смешанное произведение векторов равно определителю матрицы, составленной из этих векторов.

```python
# составим матрицу из этих векторов
arr = np.matrix([AB, AC, AD])

# найдем определитель матрицы
arr_det = np.linalg.det(arr)

# нахождение объема треугольной пирамиды
V2 = arr_det / 6
print(V2)
```

*Ответ: Объем треугольной пирамиды ABCD = 9*
