**1.	Изобразить на одном графике кривую $y=tg(x)$, заданную на $[-3/2; 3/2]$ и ее образы после поворота относительно начала координат на угол $45°$ с шагом $5°$.**

```python
def changer(x, y, alpha, a, b):
    x1 = x*np.cos(-alpha) - y*np.sin(-alpha) + a
    y1 = x*np.sin(-alpha) + y*np.cos(-alpha) + b
    return x1, y1

fig = plt.figure(figsize =(10, 10))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-3/2, 3/2, 50)
y = x
z = np.tan(x)
ax.set_xlabel('X-axis'); ax.set_xlim(-1, 1)
ax.set_ylabel('Y-axis'); ax.set_ylim(-1, 1)
ax.set_zlabel('Z-axis'); ax.set_zlim(-1, 1)

ax.plot(x, y, z)
ax.set_title("Кривая y=tg(x)")

i = 0.087
while i <= 0.79:
    x1, y1 = changer(x, y, i, 0, 0)
    ax.plot(x1, y1, z)
    i+=0.087

plt.show()
```

**2.	Построить двуполостный гиперболоид вращением гиперболы $\frac{x^2}{9}-\frac{y^2}{4}=1$ вокруг оси Ох.**

```python
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)

a = 3
b = 2

Z = np.sqrt(((a**-1)*X **2 + (b**-1)*Y **2) + 1)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(Z, Y, X, cmap='jet', alpha=0.6, vmin=-5, vmax=5)
ax.plot_surface(-Z, Y, X, cmap='jet', alpha=0.6, vmin=-5, vmax=5)

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("График двуполостного гиперболоида")

plt.show()
```

**3.	Постройте тор, используя его параметрическое уравнение:**

$\begin{equation*}
 \begin{cases}
   x=(a+b*cos(φ))*cos(θ)
   \\
   y=(a+b*cos(φ))*sin(θ)
   \\
   z=b*sin(φ)
 \end{cases}
\end{equation*}    \\0≤φ<2π;  \\0≤θ<2π;    \\b<a$

```python
import matplotlib.pyplot as plt
import numpy as np
 
a = 4
b = 1

u, v = np.mgrid[0:2 * np.pi:30j, 0:2 * np.pi:30j]
x = (a + b*np.sin(v)) * np.cos(u)
y = (a + b*np.sin(v)) * np.sin(u)
z = b*np.cos(v)
 
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap=plt.get_cmap('plasma'))

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_zlim(-2, 2)
ax.set_title("График тора")
ax.view_init(20, -40)
plt.show()
```

**4.	Построить эллиптический цилиндр.**

```python
import matplotlib.pyplot as plt 
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
 
u = np.linspace (0,2 * np.pi, 50)
h = np.linspace (0,1,20)
x = np.outer (np.sin (u), np.ones (len (h)))
y = np.outer (np.cos (u), np.ones (len (h)))
z = np.outer (np.ones (len (u)), h) 

ax.plot_surface(x, y, z, cmap=plt.get_cmap('plasma'))

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("График эллиптического цилиндра")
ax.view_init(20, -40)
plt.show()
```
