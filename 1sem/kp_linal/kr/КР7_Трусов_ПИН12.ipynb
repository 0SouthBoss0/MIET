**1.	Построить кривую $r=1-2cos(φ)$ в полярных координатах (кардиоида).**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

phi = np.linspace(-2*np.pi, 2*np.pi, 1000)
plt.polar(phi, 1-2*np.cos(phi))
plt.grid(True)
plt.show()
```

**2. Привести уравнение кривой второго порядка к каноническому виду и определить ее тип $4x^2-y^2+40x+4y+92=0$. Изобразить эту кривую и ее характеристики.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

x_s, y_s = symbols('x_s y_s')
eqq = 4*x_s**2-y_s**2+40*x_s+4*y_s+92
print(solve(eqq, y_s))
```

```python
a = np.sqrt(1)
b = np.sqrt(4)
x_center = -5
y_center = 2
c = np.sqrt(a**2+b**2)

# построение гиперболы
x = np.arange(-8, -4, 0.01)
y1 = 2 - 2*np.sqrt((x + 4)*(x + 6))
y2 = 2 + 2*np.sqrt((x + 4)*(x + 6))
plt.plot(x, y1, lw=2,color='b')
plt.plot(x, y2, lw=2,color='b')
plt.plot(-x+2*x_center, y1, lw=2,color='b')
plt.plot(-x+2*x_center, y2, lw=2,color='b')

# построение асимптот
x = np.arange(-8, -2, 0.01)
y1 = (b/a)*(x-x_center) + y_center
y2 = -(b/a)*(x-x_center) + y_center
plt.plot(x, y1, "k:")
plt.plot(x, y2, "k:")

#построение фокусов гиперболы
plt.plot(x_center-c, y_center,'ro')
plt.plot(x_center+c, y_center,'ro')

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation = 0)
plt.grid()
plt.axis("equal")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

**3. Привести уравнение кривой второго порядка к каноническому виду и определить ее тип $29x^2-24xy+36y^2+82x-96y-91=0$. Найти начало и базис соответствующей системы координат и угол её поворота относительно исходной. Изобразить эту кривую, начало координат и базисные векторы в старой и новой системах координат.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

x, y = symbols('x y')
eqq = 29*x**2-24*x*y+36*y**2+82*x-96*y-91
print(solve(eqq, y))
```

```python
x = np.arange(-3, 3, 0.01)
y1 = x/3 - np.sqrt(-25*x**2 - 50*x + 155)/6 + 4/3
y2 = x/3 - np.sqrt(-25*x**2 - 50*x + 155)/6 + 4/3
plt.plot(x, y1, lw=2,color='b')
plt.plot(x, y2, lw=2,color='b')

# параметры координатной плоскости
plt.xlabel("x")
plt.ylabel("y", rotation = 0)
plt.grid()
plt.axis("equal")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

**4.	Построить кривую и определить ее тип.**

$\begin{equation*}
 \begin{cases}
   x = ±2\sqrt{t}
   \\
   y=3
   \\
   z=4t
 \end{cases}
\end{equation*}$
