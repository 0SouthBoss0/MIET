**2. Найти приближённые и точные значения в алгебраической форме корней $\sqrt[3] {27i}$. Используя операции отношения для приближённых значений корней  выяснить:**

**1) какие из найденных корней изображаются на комплексной плоскости точками, лежащими на мнимой оси;**

**2) какой из найденных корней имеет главный аргумент, равный 5pi/6.**

```python
import matplotlib.pyplot as plt
import numpy as np
import math

z = 27j
stepen = 3

r = np.abs(z)
phi=np.angle(z)
k = np.arange(1,stepen+1)

zroot = r**(1/stepen)*(np.cos((phi+2.0*np.pi*k)/stepen) + 1j*np.sin((phi+2.0*np.pi*k)/stepen))

print("Точные значения корней:", *zroot)
print("Приближённые значения корней:", *np.round(zroot))

plt.title('$\sqrt[3]{27i}$ complex roots')

plt.plot(np.real(zroot), np.imag(zroot), '*g')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

t = np.linspace(0,2*np.pi,100)
plt.plot(r**(1/stepen)*np.cos(t),r**(1/stepen)*np.sin(t),':')

x1=np.real(zroot[0])
y1=np.imag(zroot[0])

x2=np.real(zroot[1])
y2=np.imag(zroot[1])

x3=np.real(zroot[2])
y3=np.imag(zroot[2])

plt.plot([x1, x2],[y1, y2], '-.b')
plt.plot([x2, x3],[y2, y3], '-.b')
plt.plot([x3, x1],[y3, y1], '-.b')

plt.xlabel('Re(z)')
plt.ylabel('Im(z)', rotation = 0)

plt.grid(True)

# одинаковый масштаб по осям
plt.axis('scaled')

plt.show()

print("\n1) На мнимой оси:")
for i in range(len(zroot)):
    if np.isclose(zroot[i].real, 0)==1:
        print(zroot[i])

print("\n3) Аргумент равен 5pi/6:")
for i in range(len(zroot)):
    if np.isclose(np.angle(zroot[i]), 5*np.pi/6)==1:
        print(zroot[i])
```

Для последовательности ${x_n}= \frac{\sqrt{2n^3 + 2n+1}}{\sqrt{n^3-2n}}$ выполнить: <br>

1.Найти a = $\lim\limits_{x\to\infty}$ $x_n$.<br>
2.Вычислить $n_0(0.01)$, $n_0(0.001)$.

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from sympy.abc import n

#############################
n0 = 10
x_n = [(sp.sqrt(2*n**3+2*n+1)) / (sp.sqrt(n**3-2*n)) for n in range(n0-2, n0+10+1)]
a = sp.limit( (sp.sqrt(2*n**3+2*n+1)) / (sp.sqrt(n**3-2*n)) , n, sp.oo)
eps = 0.01

plt.subplot(1,2,1)

print("Предел:", a)

x = [n for n in range(n0-2, n0+10+1)]
y = x_n
plt.plot(x,y, 'm:*')

x = [n for n in range(n0-2, n0+10+1)]
y = [a-eps for n in range(n0-2, n0+10+1)]
plt.plot(x,y, 'k-')

x = [n for n in range(n0-2, n0+10+1)]
y = [a+eps for n in range(n0-2, n0+10+1)]
plt.plot(x,y, 'k-')

a = sp.solve_univariate_inequality(sp.Abs( (sp.sqrt(2*n**3+2*n+1)) / (sp.sqrt(n**3-2*n)) - a )<eps,n,relational=False)
b = sp.solve_univariate_inequality(n>0,n,relational=False)

print("n0 (", eps, ") = ", sp.Intersection(a,b), sep="")

# нанести координатную сетку
plt.grid()


################################
n0 = 40
x_n = [(sp.sqrt(2*n**3+2*n+1)) / (sp.sqrt(n**3-2*n)) for n in range(n0-2, n0+10+1)]
a = sp.limit( (sp.sqrt(2*n**3+2*n+1)) / (sp.sqrt(n**3-2*n)) , n, sp.oo)
eps = 0.001

plt.subplot(1,2,2)

x = [n for n in range(n0-2, n0+10+1)]
y = x_n
plt.plot(x,y, 'm:*')

x = [n for n in range(n0-2, n0+10+1)]
y = [a-eps for n in range(n0-2, n0+10+1)]
plt.plot(x,y, 'k-')

x = [n for n in range(n0-2, n0+10+1)]
y = [a+eps for n in range(n0-2, n0+10+1)]
plt.plot(x,y, 'k-')

a = sp.solve_univariate_inequality(sp.Abs( (sp.sqrt(2*n**3+2*n+1)) / (sp.sqrt(n**3-2*n)) - a )<eps,n,relational=False)
b = sp.solve_univariate_inequality(n>0,n,relational=False)

print("n0 (", eps, ") = ", sp.Intersection(a,b), sep="")

# нанести координатную сетку
plt.grid()

plt.show()
```
