**Упражнение 4. В одной системе координат изобразить векторами разного цвета числа. Нанести координатную сетку, отобразить оси линиями черного цвета, подписать их. Масштаб по осям сделать одинаковым. Подписать графическое окно. Прокомментировать геометрический смысл суммы и разности комплексных чисел.**

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([1+3j, 3+4j])
x2 = np.array([(1+3j)+(3+4j), (1+3j)-(3+4j)])
orig = np.zeros(2)

plt.plot()
q = plt.quiver(orig,orig, np.real(x1), np.imag(x1), angles='xy', scale_units='xy', scale=1, color="green")
q = plt.quiver(orig,orig, np.real(x2), np.imag(x2), angles='xy', scale_units='xy', scale=1, color="blue")

plt.axis([-2,7,-2,7])

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.grid(True)

plt.title('Упражнение 4')

plt.xlabel('Re(z)')
plt.ylabel('Im(z)', rotation = 0)

plt.show()
```

*Сумма комплексных чисел представляется суммой векторов. Разность комплексных чисел представляется разностью векторов.*

**Упражнение 5. Найти и изобразить точками на комплексной плоскости все корни. Изобразить пунктиром окружность, на которой эти точки лежат. Построить штрих-пунктиром правильный многоугольник с вершинами в этих точках. Нанести сетку, отобразить оси линиями черного цвета, подписать их. Масштаб по осям сделать одинаковым. Подписать графическое окно.**

```python
import matplotlib.pyplot as plt
import numpy as np
import math

z = 2*math.sqrt(3)-2j

r = np.abs(z)
phi=np.angle(z)
k = np.arange(1,5)

zroot = r**(1/4)*( np.cos((phi+2.0*np.pi*k)/4) + 1j*np.sin((phi+2.0*np.pi*k)/4))

print("Корни:", *zroot)
plt.title('$\sqrt[4]{2\sqrt{3}-2i}$ complex roots')

plt.plot(np.real(zroot), np.imag(zroot), '*g')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

t = np.linspace(0,2*np.pi,100)
plt.plot(r**(1/4)*np.cos(t),r**(1/4)*np.sin(t),':')

x1=np.real(zroot[0])
y1=np.imag(zroot[0])

x2=np.real(zroot[1])
y2=np.imag(zroot[1])

x3=np.real(zroot[2])
y3=np.imag(zroot[2])

x4=np.real(zroot[3])
y4=np.imag(zroot[3])

plt.plot([x1, x2],[y1, y2], '-.b')
plt.plot([x2, x3],[y2, y3], '-.b')
plt.plot([x3, x4],[y3, y4], '-.b')
plt.plot([x4, x1],[y4, y1], '-.b')

plt.xlabel('Re(z)')
plt.ylabel('Im(z)', rotation = 0)

plt.grid(True)

# одинаковый масштаб по осям
plt.axis('scaled')

plt.show()
```

**Контрольные вопросы:**

**1) Каким образом можно ввести комплексное число?**

*Для ввода комплексного числа можно пользоваться встроенной функцией complex (a,b), где a - действительная, b – мнимая часть вводимого комплексного числа. Также можно использовать конструкции a+bj, в которых мнимые части комплексных чисел сопровождаются буквой j.*

**2) Что такое вещественная и мнимая часть комплексного числа?**

*Действительное число a называется действительной частью комплексного числа z=a+bi и обозначается a=Rez.*

*Действительное число b называется мнимой частью числа z=a+bi и обозначается b=Imz.*

**3) Как осуществляются арифметические действия над комплексными числами?**

*Сложение, вычитание, умножение и возведение в степень комплексных чисел осуществляется с помощью тех же встроенных функций и с использованием тех же символов, которые служат для аналогичных операций над действительными числами.*

**4) Что такое модуль и аргумент комплексного числа?**

*Длина вектора, изображающего комплексное число, называется модулем комплексного числа.*

*Угол между осью абсцисс и вектором OM, изображающим комплексное число a + b·i, называется аргументом комплексного числа a + b·i*

**5) Чему равны модуль и аргумент произведения и частного комплексных чисел?**

*Модуль произведения двух комплексных чисел равен произведению их модулей, а аргумент — сумме их аргументов.*

*Модуль частного двух комплексных чисел равен частному модулей их модулей, а аргумент - разности их аргументов.*

**6) Сколько существует корней n-ой степени из комплексного числа? Как они расположены на комплексной плоскости?**

*N корней из комплексного числа N-степени, они расположены в вершинах правильного N-угольника.*
