**Упражнение 1. Построить графики функций $y1 = \frac{2x-1}{x-1}$, $y2 = \frac{2x-1}{|{x-1}|}$ на таких промежутках, чтобы можно было судить о поведении этих функций.**

```python
import sympy as sp
from matplotlib import style
from sympy.abc import x
import matplotlib.pyplot as plt

style.use('ggplot') # отображение сетки координат по умолчанию
y = (2*x-1)/(x-1)
sp.plot((y,(x,-3,0.9)),(y,(x,1.1,3)),adaptive=False,nb_of_points=1000,line_color='r')
plt.show()
```

**Упражнение 2. Используя графики функций, найдите приближенно пределы (или убедитесь, что они не существуют):**

$а)\lim\limits_{x\to +\infty}(1+x)^\frac{1}{x} $ 

```python
import sympy as sp
from matplotlib import style
from sympy.abc import x
import matplotlib.pyplot as plt

y = (1+x)**(1/x)
style.use('ggplot') # отображение сетки координат по умолчанию
sp.plot((y,(x,5,-0.01)),(y,(x,0.01,100)),adaptive=False,nb_of_points=100,line_color='r')
plt.show()
```

*По полученному графику можно предположить, что предел равен 1.*

$а)\lim\limits_{x\to 0}\frac{\sin{\frac{x^2}{\sqrt{1-x^3}}  }}{x\ln(1-x^3)} $ 

```python
import sympy as sp
from matplotlib import style
from sympy.abc import x
import matplotlib.pyplot as plt

y = (  (sp.sin((x**2)/(sp.sqrt(1-x**3)))) / (x*sp.ln(1-x**3)) )
style.use('ggplot') # отображение сетки координат по умолчанию
sp.plot((y,(x,-3,-0.1)),(y,(x,0.1,3)),adaptive=False,nb_of_points=100,line_color='r')
plt.show()
```

*По полученному графику можно предположить, что предел равен -∞.*

**Упражнение 3. Вычислите точные значения пределов из упражнения 2 . Соответствуют ли они полученным приближениям?**

```python
from sympy.abc import x
print("а)", sp.limit((1+x)**(1/x),x,sp.oo))
print("б)", sp.limit(((sp.sin((x**2)/(sp.sqrt(1-x**3)))) / (x*sp.ln(1-x**3))),x,0))
```

*Точные значения пределов соответствуют полученным приближениям.*

**Упражнение C2. Используя графики функций, найдите приближенно пределы (или убедитесь, что они не существуют):**

$а)\lim\limits_{x\to \infty} \ln(x) \sin{x} $ 

```python
import sympy as sp
from matplotlib import style
from sympy.abc import x
import matplotlib.pyplot as plt

y = sp.ln(x)*sp.sin(x)
style.use('ggplot') # отображение сетки координат по умолчанию
sp.plot(y,(x,-50,50),adaptive=False,nb_of_points=1000,line_color='r')
plt.show()
```

*По полученному графику можно предположить, что предела нет.*

$б)\lim\limits_{n\to +\infty}\frac{2^n}{n!} $ 

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(1, 25, 25)
x = [int(x[i]) for i in range(len(x))]
y = [(2**x[i])/(sp.factorial(x[i])) for i in range(len(x)) ]

plt.plot(x,y)
plt.xlabel("x"); plt.ylabel("y", rotation = 0)
plt.grid(True)
plt.show()
```

*По полученному графику можно предположить, что предел равен 0.*
