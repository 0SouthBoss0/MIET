**Упражнение 3.
Используя команду plt.legend, в одной системе координат построить графики
функций, подписать оси, для каждого графика задать цвет, тип линии и форму
маркера и сделать для них общую легенду-справку.**

```python
import matplotlib.pyplot as plt
import numpy as np

# область значений x для расчёта графика
x = np.linspace(-2*np.pi,2*np.pi,1000)

# график y(x) = cos(x)
y = np.cos(x)
handle1, = plt.plot(x,y, 'm:*', label='y=cos(x)')

# график y(x) = 2cos(x)
y = 2*np.cos(x)
handle2, = plt.plot(x,y, 'b-.+', label='y=2cos(x)')

# график y(x) = 0,3cos(x)
y = 0.3*np.cos(x)
handle3, = plt.plot(x,y, 'g-x', label='y=0,3cos(x)')

# график y(x) = -cos(x)
y = -np.cos(x)
handle4, = plt.plot(x,y, 'k-p', label='y=-cos(x)')

# подписать оси
plt.xlabel('x')
plt.ylabel('y(x)', rotation = 0)

# отобразить оси координатной плоскости
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

# подписать график
plt.title('Упражнение 3')

# нанести координатную сетку
plt.grid()

# вывести легенду графиков
plt.legend(handles=[handle1,handle2,handle3, handle4])

# показать график
plt.show()
```

**Упражнение 4. Преобразование графиков функций.**

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-5,5,1000)

plt.subplot(2,3,1)
plt.plot(x,abs(abs(x)-2), 'y')
plt.grid(True)

plt.subplot(2,3,2)
plt.plot(x,abs(abs(x-2)-2),'m')
plt.grid(True)

plt.subplot(2,3,3)
plt.plot(x,abs(abs(x+2)-2),'c')
plt.grid(True)

plt.subplot(2,3,4)
plt.plot(x,abs(abs(2*x)-2),'g')
plt.grid(True)

plt.subplot(2,3,5)
plt.plot(x,abs(abs(0.5*x)-2),'b')
plt.grid(True)

plt.subplot(2,3,6)
plt.plot(x,abs(abs(-x)-2),'k')
plt.grid(True)

plt.show()
```

4) График y=f(2x) получается из графика y=f(x) сжатием в 2 раза вдоль оси абсцисс.

**Упражнение 6.
Задайте массив значений переменной x = 0.001, 0.01, 0.1, 1, 10, 100, 1000.**

```python
import matplotlib.pyplot as plt
import numpy as np

# массив степеней
x = np.arange(-3,4,1) 

# массив вида [10 10 ... 10]
z = np.ones((7))*10

# поэлементное возведение в степень
y = np.power(z,x)

fig = plt.figure()

# график в линейной шкале
ax1 = fig.add_subplot(121)
ax1.plot(x,y)

# подписать график
ax1.title.set_text('Линейная шкала')

# график в полулогарифмической шкале
ax2 = fig.add_subplot(122)
ax2.semilogy(x,y)

# подписать график
ax2.title.set_text('Полулогарифмическая шкала')

# нанести координатную сетку
ax1.grid()
ax2.grid()

# показать график
plt.show()
```

**Упражнение 7.
В одной системе координат построить графики функции, и
график обратной функции. Первый график построить пунктирной линией зеленого
цвета, второй – штрих-пунктирной линией красного цвета. Отобразить оси в виде
сплошных линий черного цвета. Масштаб по осям сделать одинаковым с помощью
команд plt.axis('equal'). Подписать оси, нанести координатную сетку.
Построить прямую, относительно которой графики симметричны (в виде сплошной
линии синего цвета). Вывести заголовок.**

```python
import matplotlib.pyplot as plt
import numpy as np

# график y(x) = cos(x)
x = np.linspace(0, np.pi, 100)
y = np.cos(x)
handle1, = plt.plot(x,y, 'g:', label='y=cos(x)')

# график y(x) = arccos(x)
x = np.linspace(0, 1, 1000)
y = np.arccos(x)
handle2, = plt.plot(x,y, 'r-.', label='y=1/cos(x)')

# прямая симметрии графиков y(x)=x 
x = np.linspace(0, np.pi, 1000)
y = x
handle3, = plt.plot(x, y, 'b-', label='y=x')

# отображение осей
plt.axhline(y=0, color='k')  
plt.axvline(x=0, color='k')

# подписать оси
plt.xlabel('x')
plt.ylabel('y', rotation = 0)

# одинаковый масштаб по осям
plt.axis('equal')

# нанести координатную сетку
plt.grid()

# вывести легенду графиков
plt.legend(handles=[handle1,handle2,handle3])

# установить заголовок
plt.title("Упражнение 7")
          
# показать график
plt.show()
```
